```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClusteringmerativeClustering
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

```python
x = df.iloc[:, [3, 4]].values
plt.figure(figsize=(12,8))
dendro = shc.dendrogram(shc.linkage(x, method="ward"))
ax = plt.gca()
ax.yaxis.grid(True, linestyle="--")
plt.title("Dendrogram plot")
plt.xlabel("Euclidean Distances")
plt.ylabel("Customers")
plt.show()
```

```python
clusters = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_pred = clusters.fit_predict(x)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x[y_pred == 0, 0], x[y_pred == 0, 1], s = 50, c = 'cyan', label = 'C1')
plt.scatter(x[y_pred == 1, 0], x[y_pred == 1, 1], s = 50, c = 'red', label = 'C2')
plt.scatter(x[y_pred == 2, 0], x[y_pred == 2, 1], s = 50, c = 'green', label = 'C3')
plt.scatter(x[y_pred == 3, 0], x[y_pred == 3, 1], s = 50, c = 'black', label = 'C4')
plt.scatter(x[y_pred == 4, 0], x[y_pred == 4, 1], s = 50, c = 'violet', label = 'C5')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
