```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import figure
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn import preprocessing
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
label_encoder = preprocessing.LabelEncoder()
df['Genre'] = label_encoder.fit_transform(df['Genre'])
df['Genre'].head()
```

```python
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])
plt.xlabel('Loan Amount')
plt.ylabel('Applicant Income')
```

```python
wcss_values = []
income_score = df[['Annual Income (k$)', 'Spending Score (1-100)']]
for i in range(1,7):
    kmeans  = KMeans(i)
    kmeans.fit(income_score.iloc[:,0:2])
    wcss_iter = kmeans.inertia_
    wcss_values.append(wcss_iter)

number_clusters = range(1, 7)
plt.plot(number_clusters, wcss_values)
plt.title('Elbow method')
plt.xlabel("NUmber of Clusters")
plt.ylabel('WCSS value')
plt.show()
```

```python
kmeans  = KMeans(5)
income_score = df[['Annual Income (k$)', 'Spending Score (1-100)']]
income_score

identified_clusters = kmeans.fit_predict(income_score)
income_score["clusters"] = identified_clusters

plt.scatter(income_score["Annual Income (k$)"], income_score["Spending Score (1-100)"], c=income_score["clusters"], cmap='rainbow')
plt.ylabel('Spending Score')
plt.xlabel('Applicant Income')
plt.show()
```
